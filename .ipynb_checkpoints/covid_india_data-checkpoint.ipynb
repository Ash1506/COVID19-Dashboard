# Importing the libraries

```python
# import libraries
# ================

# for date and time opeations
from datetime import datetime
# for file and folder operations
import os
# for regular expression opeations
import re
# for listing files in a folder
import glob
# for getting web contents
import requests 
# storing and analysing data
import pandas as pd
# for scraping web contents
from bs4 import BeautifulSoup
# regular expression
import re
# for numerical analysis
import numpy as np
```

# Parsing the data

```python
link = 'https://www.mohfw.gov.in/'
# get web data
req = requests.get(link)
# parse web data
soup = BeautifulSoup(req.content, "html.parser")
```

```python
thead = soup.find_all('thead')[-1]
```

```python
print(soup)
```

# Collecting the table headings

```python
print(thead)
```

```python
head=thead.find_all('tr')
print(head)
```

```python
tbody=soup.find_all('tbody')[-1]
#print(tbody)
body=tbody.find_all('tr')
#print(body)
```

# Getting the table content

```python
head_rows=[]
body_rows=[]
for tr in head:
    td = tr.find_all(['th', 'td'])      #Collecting the headings in the list
    row = [i.text for i in td]
    head_rows.append(row)
#print(head_rows)    
```

```python
for tr in body:
    td = tr.find_all(['th', 'td'])      #Collecting the state-wise data in the list
    row = [i.text for i in td]
    body_rows.append(row)
print(body_rows)    
```

```python
df_bs=pd.DataFrame(body_rows[:len(body_rows)-6],columns=head_rows[0])   #Putting the above list in a dataframe
df_bs.drop('S. No.', axis=1,inplace=True)
df_bs.head(36)         
```

# Adding the date column 

```python
now=datetime.now()
df_bs['Date(mm/dd/yyyy)']=now.strftime("%m/%d/%Y")
df_bs
```

```python
df_bs['Name of State / UT']=df_bs['Name of State / UT'].str.replace('#','')   #Removing the extra characters
df_bs
```

# Saving the data in a csv file 

```python
df_bs.rename(columns={"Cured/Discharged/Migrated*":"Cured*"},inplace=True)
df_bs
df_bs.to_csv('india_data.csv',index=False)
```
